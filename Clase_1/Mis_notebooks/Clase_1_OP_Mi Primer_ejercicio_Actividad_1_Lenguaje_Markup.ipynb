Creamos una lista ramdom

```python
# Importamos la libreria random
import random
lista_estatica = []
lista_random = list()

for i in range(6):
  lista_estatica.append(i)

for i in range(6):
    lista_random.append(random.randint(1,10))

print("Mi lista estática", lista_estatica)
print("Mi lista random", lista_random)
```
